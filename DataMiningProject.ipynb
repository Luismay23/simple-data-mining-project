```python
from sklearn.linear_model import LinearRegression
```

```python
#Model 1 :
```

```python
lm = LinearRegression()
lm
```

```python
df['SAT_Math_Score'] = df['SAT_Math_Score'].astype(str).str.replace(',', '', regex=False).astype(float)
df['No_AP_TestTakers'] = df['No_AP_TestTakers'].astype(str).str.replace(',', '', regex=False).astype(float)
df['Total_Exam_Taken'] = df['Total_Exam_Taken'].astype(str).str.replace(',', '', regex=False).astype(float)
```

```python
#For the above lines, I did them because of an error : 'ValueError: could not convert string to float: '1,510' when I tried to fit X and Y
#The lines above convert the values to strings, then removes the commas, then converts them to float. This way the lm.fit(X,Y) works
```

```python
Y = df['SAT_Math_Score']
X = df[['No_AP_TestTakers', 'Total_Exam_Taken']]
```

```python
print(df[['No_AP_TestTakers', 'Total_Exam_Taken', 'SAT_Math_Score']].dtypes)
```

```python
lm.fit(X,Y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
Yprt = lm.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y,Yprt)
```

```python
#Model 2 :
```

```python
df['No_Exam_Passed'] = df['No_Exam_Passed'].astype(str).str.replace(',', '', regex=False).astype(float)
```

```python
Y2 = df['SAT_Math_Score']
X2 = df[['No_AP_TestTakers', 'Total_Exam_Taken', 'No_Exam_Passed']]
```

```python
lm2 = LinearRegression()
lm2
lm2.fit(X2,Y2)
```

```python
lm2.intercept_
```

```python
lm2.coef_
```

```python
Y2prt = lm2.predict(X2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y2,Y2prt)
```

```python
#Set 4 : The model using the 3 independent variables is the better variable because it's MSE (3128) was lower than that
#of the model using only the 'No_AP_TestTakers' and 'Total_Exam_Taken' (3565)
```
